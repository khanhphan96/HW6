```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import requests 
import json 
import random 
import urllib 

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
#!pip install citipy
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
from api_keys import api_key
api_key
```

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
city
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key


weather = []

print("Beginning Data Retrieval")
record_count = 1
set_count = 1


for i, city in enumerate(cities):
    if (i > 0):
        set_count += 1
        record_count += 1

    city_url = url + "&q=" + urllib.request.pathname2url(city)
    
    print(f"Processing record for {set_count} of {record_count} | {city}")
    

    try:
        request = requests.get(city_url).json()
        cloudiness = request["clouds"]["all"]
        country = request["sys"]["country"]
        date = request["dt"]
        humidity = request["main"]["humidity"]
        latitude = request["coord"]["lat"]
        longitude = request["coord"]["lon"]
        max_temp = request["main"]["temp_max"]
        wind = request["wind"]["speed"]
        
        
        weather.append({"City": city, "Latitude": latitude, "Longitude": longitude, 
                        "Max Temperature": max_temp, "Humidity": humidity, 
                  "Cloudiness": cloudiness, "Wind Speed": wind, "Country": country, "Date": date})

    except:
        print("City not found. Skipping")
        pass

print("Data Retrieval Complete")
```

#### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_plot = pd.DataFrame(weather)
cloudiness = city_plot["Cloudiness"]
humidity = city_plot["Humidity"]
latitude = city_plot["Latitude"]
max_temperature = city_plot["Max Temperature"]
wind_speed = city_plot["Wind Speed"]

city_plot.head()
```
